```python
import cv2
import numpy as np
from tensorflow.keras.models import load_model
from tensorflow.keras.applications.mobilenet_v2 import preprocess_input

# Load your trained model
model = load_model("mask_detector_model.keras")

# Simple built-in face detector
face_cascade = cv2.CascadeClassifier(
    cv2.data.haarcascades + "haarcascade_frontalface_default.xml"
)

labels = ["Mask", "No Mask"]

cap = cv2.VideoCapture(r"C:\Users\elmoa\Downloads\Video_Generation_for_CNN_Testing.mp4")  

while True:
    ret, frame = cap.read()
    if not ret:
        break

    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # detect faces
    faces = face_cascade.detectMultiScale(gray, 0.5, 3)

    for (x, y, w, h) in faces:

        if w < 50 or h < 50:  # ignore very small detections
                continue
        face = frame[y:y+h, x:x+w]

        # preprocess face for MobileNetV2
        face_resized = cv2.resize(face, (224, 224))
        face_rgb = cv2.cvtColor(face_resized, cv2.COLOR_BGR2RGB)
        face_input = preprocess_input(face_rgb)
        face_input = np.expand_dims(face_input, axis=0)

        # predict
        pred = model.predict(face_input)[0]
        class_id = np.argmax(pred)
        confidence = pred[class_id]

        # label + color
        color = (0, 255, 0) if class_id == 0 else (0, 0, 255)
        text = f"{labels[class_id]} ({confidence:.2f})"

        # draw box
        cv2.rectangle(frame, (x, y), (x+w, y+h), color, 2)
        cv2.putText(frame, text, (x, y - 10),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.8, color, 2)

    cv2.imshow("Mask Detection", frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
